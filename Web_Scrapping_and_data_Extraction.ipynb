```python
# Download Libraries
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
# Initialising URL and directory
base_url="https://clutch.co/web-developers?page="
Comp_Database=[]

# Find Last page number
response_1=requests.get(base_url+'0')
soup_1=bs(response_1.content,'html.parser')
last_url=soup_1.find('li',class_="page-item last")
last_page=int(last_url.find('a')['data-page'])+1
print(" Last Page is : ", last_page)
```

```python
# Defining data collection function
def dataprep(url_name):
  response=requests.get(url_name)
  soup=bs(response.content,'html.parser')
  raw_data=soup.find("ul","directory-list shortlist")
  data_list=raw_data.find_all('li', attrs= {'class':"provider provider-row"})
  for row in data_list:
    Comp_Data={}
    Comp_Data['Name']=row['data-title'] 
    #.find('li',class_="provider provider-row")['data-title']
    try:
      Comp_Data['Country']= str(row.findAll(class_="locality")).split(",")[1].split("</")[0].strip()
    except:
      Comp_Data['Country']='NaN'
    Comp_Data['Website']= row.find('a', class_="website-link__item")['href']
    Services=row.findAll('div', class_="carousel-item")
    j=0
    for service in Services:
      Comp_Data["Service_"+str(j)]=(" ".join(service.text.split(' ')[1:]))
      Comp_Data["Service_"+str(j)+" Share"]=int(service.text.split(' ')[0].split('%')[0])
      #print(Comp_Data["Service_"+str(j)])
      #print(j)
      j=j+1
    Comp_Database.append(Comp_Data)
```

```python
#Scarping through all the web pages
for i in range(last_page):
  url= base_url+str(i)
  print(url)
  dataprep(url)
  i=+1
```

```python

# Saving raw file to CSV
df=pd.DataFrame(Comp_Database)
df.to_csv('raw.csv')
```
